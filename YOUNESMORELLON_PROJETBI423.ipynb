```python
#Projet BI423: D√©tection d'oriC - Code

#Biblioth√®ques
import re
import matplotlib.pyplot as plt
import customtkinter as ctk
from tkinter import filedialog
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure


#Etape 1 : Charger la s√©quence √† partir du fichier FASTA
def lire_sequence_fasta(nom_fichier):
    """Lit une s√©quence FASTA (ignore la premi√®re ligne de description)"""
    sequence = ""
    with open(nom_fichier, 'r') as fichier:
        #lit la premi√®re ligne et l'ignore
        fichier.readline()
        ligne = fichier.readline()
        while ligne:
            sequence += ligne.strip()  #enlever les espaces et les sauts de ligne
            ligne = fichier.readline()  #lire la ligne suivante
    return sequence.upper()


#Etape 2 : Rechercher les r√©gions riches en A et T
def calcul_regions_AT(sequence):
    """Renvoie les r√©gions riches en A/T en comparant le nombre de G/C au nombre de A/T dans une r√©gion"""
    taille_r√©gion = 300
    r√©gions_dint√©r√™t = [] #liste pour stocker les r√©gions s√©lectionn√©es
    #parcours de la s√©quence par tranches de 300 bases
    for i in range(0, len(sequence) - taille_r√©gion + 1, taille_r√©gion):
        region = sequence[i:i+taille_r√©gion] #extraction de la sous-s√©quence
        cpt_AT = 0 #compteur A/T
        cpt_CG = 0 #compteur C/G
        #comptage des types de bases dans la r√©gion
        for base in region:
            if base == 'A' or base == 'T':
                cpt_AT += 1
            elif base == 'C' or base == 'G':
                cpt_CG += 1
        #v√©rifie si la r√©gion est riche en A/T
        if cpt_AT >= cpt_CG:
            r√©gions_dint√©r√™t.append((i, i + taille_r√©gion, region)) #ajoute les positions de d√©but et de fin de r√©gion + la r√©gion

    return r√©gions_dint√©r√™t


#Etape 3 : D√©tecter les motifs DnaA
def motifs_consensus(regions):
    """Renvoie les motifs consensus trouv√© dans la s√©quence s'il y en a"""
    motifs = [
        r"TTATCCACA", r"TTGTCCACA", r"TTATGCACA", r"TTATCCAGA", r"TTATTCACA",
        r"TTATCCAAA", r"TCATTCACA", r"TCATTGACA", r"TTGTGCACA", r"ATATTTAAA"
    ]
    r√©sultats = []
    for debut, fin, region in regions: #on cherche des motifs dans les s√©quences riches en AT
        for motif in motifs: 
            for match in re.finditer(motif, region): #extraire et localiser les motifs
                position_globale = debut + match.start()  #calcul de la position dans la s√©quence globale
                motif_trouv√© = match.group() #motif trouv√© dans la r√©gion
                r√©sultats.append((position_globale, motif_trouv√©))

    return r√©sultats


# √âtape 4 : Calculer le GC-skew
def GC_skew(seq):
    """Calcule le GC-skew d'une s√©quence"""
    C = seq.count('C')
    G = seq.count('G')
    
    somme = G+C

    if somme == 0: 
        return 0 #Pour √©viter de faire une division par 0
     
    return (G-C) / somme #calcul du GC-skew

def GCskewfenetre (seq, fenetre):
    """Calcule l'asym√©trie GC dans une fen√™tre"""
    sequences = [ ]
    position = [ ]
    for i in range (0, len(seq) - fenetre + 1 , fenetre): #Pour calculer fenetre par fenetre
        seq_fenetre = seq[i:i+fenetre]
        val = GC_skew(seq_fenetre)
        sequences.append(val) 
        position.append(i + fenetre//2) # "fenetre//2" est utilis√© pour placer chaque valeur de GC_skew au milieu de sa fen√™tre (pour mieux visualiser sur le graphe)
    return position, sequences



#Etape 5 : D√©tecter l'OriC bas√© sur le GC-skew
def detect_oriC(positions, sequences, seuil=0.1):
    """D√©tecte l'oriC en cherchant un changement de signe dans le GC-skew."""
    oriC_positions = []
    for i in range(1, len(sequences)):
        if abs(sequences[i] - sequences[i-1]) > seuil and sequences[i-1] < 0 and sequences[i] > 0:  # D√©tection d'un passage de n√©gatif √† positif et d√©passement d'un seuil
            oriC_positions.append(positions[i])  # Enregistre la position du changement
    return oriC_positions


# ---- Programme principal ----

# √âtape 1 : Lire la s√©quence
seq = lire_sequence_fasta('sequenceprojet.txt')

# √âtape 2 : D√©tecter les r√©gions riches en AT
regions = calcul_regions_AT(seq)

# √âtape 3 : D√©tecter les motifs DnaA dans toute la s√©quence (comme la fonction les cherche dans toute la s√©quence)
motifs = motifs_consensus(regions)
print("Motifs DnaA trouv√©s :", motifs)

# √âtape 4 : Calculer le GC-skew
fenetre = 5000  # Taille de la fen√™tre pour le calcul du GC-skew
pos_gc, val_gc = GCskewfenetre(seq, fenetre) 

# √âtape 5 : Filtrer les OriC d√©tect√©s
oriC_pos = detect_oriC(pos_gc, val_gc)

# Partie affichage du graphe
plt.figure(figsize=(10, 5))
plt.plot(pos_gc, val_gc, marker='o')
plt.title("GC-skew avec fen√™tre de taille " + str(fenetre))
plt.xlabel("Position")
plt.ylabel("GC-skew")
plt.axhline(0, color='gray', linestyle='--')  # tracer une ligne horizontale √† y = 0 pour visualiser o√π le GC-skew change de signe

# Marquer les positions de l'OriC d√©tect√©es sur le graphique
for i in oriC_pos:
    plt.axvline(x=i, color='red', linestyle='--')

plt.grid(True)
plt.show()

# Affichage des positions des OriC filtr√©es
if oriC_pos:
    print(f"OriC d√©tect√© √† la position(s) : {oriC_pos}")
else:
    print("Aucun OriC d√©tect√©.")


# ----- Interface -----

def charger_fichier(context):
    """Charge un fichier et lit la s√©quence"""
    filepath = filedialog.askopenfilename(filetypes=[("FASTA / TXT", "*.fasta *.txt")])
    context["sequence"] = lire_sequence_fasta(filepath) #lit la s√©quence
    context["fenetre_resultats"].insert("end", "Fichier charg√© avec succ√®s\n\n") #affiche que le fichier a bien √©t√© charg√©

def analyser(context):
    """Analyse la s√©quence et affiche les r√©sultats dans le graphe"""
    sequence = context.get("sequence") 
    fenetre_resultats = context["fenetre_resultats"] #r√©cup√®re la s√©quence depuis le contexte
    fenetre_resultats = context["fenetre_resultats"]  #r√©cup√®re la zone d‚Äôaffichage des r√©sultats 
    fenetre_resultats.delete("1.0", "end") # efface le contenu pr√©c√©dent dans la zone de texte

    regions = calcul_regions_AT(sequence) #recherche des r√©gions riches en A/T
    motifs = motifs_consensus(regions) #recherche des motifs consensus
    pos_gc, val_gc = GCskewfenetre(sequence, 5000) #calcul du GCskew avec une fen√™tre de 5000
    oriC = detect_oriC(pos_gc, val_gc) #d√©tection de l'origine de r√©plication

    fenetre_resultats.insert("end", f"Motifs DnaA trouv√©s : {len(motifs)}\n") #affichage des motifs trouv√©s dans la zone de r√©sultats
    for i, motif in motifs:
        fenetre_resultats.insert("end", f"  ‚Üí {motif} √† la position {i}\n")
    fenetre_resultats.insert("end", f"\nOriC d√©tect√© √† : {oriC if oriC else 'Aucun'}\n") #affiche l'ori trouv√© et sa position

    if context.get("canvas"):
        context["canvas"].get_tk_widget().destroy() #supprime ancien graphique s'il y en a un

    fig = Figure(figsize=(8, 3.5), dpi=100) #cr√©ation de la figure matplotlib
    graphe = fig.add_subplot(111) #ajoute le graphe dans la figure
    graphe.plot(pos_gc, val_gc, marker='o') #cr√©ation du graphe
    graphe.axhline(0, color='gray', linestyle='--')
    for i in oriC:
        graphe.axvline(x=i, color='red', linestyle='--')
    graphe.set_title("GC-skew avec fen√™tre de taille " + str(fenetre)) #titre du graphe
    graphe.set_xlabel("Position") #titre de l'axe x
    graphe.set_ylabel("GC-skew") #titre de l'axe y
    graphe.grid(True)

    canvas = FigureCanvasTkAgg(fig, master=context["fenetre"]) #associe le graphe √† la fen√™tre tkinter
    canvas.draw() #ajout du graphe
    canvas.get_tk_widget().pack(pady=10, fill="both", expand=True) #affiche le graphe
    context["canvas"] = canvas #enregistre la fen√™tre

#th√®mes de l'interface
ctk.set_appearance_mode("light")
ctk.set_default_color_theme("green")

def interface():
    context = {"sequence": None, "canvas": None,} #stocke la s√©quence et la figure (vide au d√©but)

    fenetre = ctk.CTk() #cr√©ation de la fen√™tre principale
    context["fenetre"] = fenetre

    fenetre.title("Recherche OriC chez une bact√©rie") #titre de l'interface
    fenetre.geometry("900x700") #taille de l'interface

    titre = ctk.CTkLabel(fenetre, text="Recherche OriC chez une bact√©rie", font=ctk.CTkFont(size=20, weight="bold")) #ajout du titre en haut de la fen√™tre
    titre.pack(pady=20)

    #bouton pour charger le fichier 
    bouton_charger = ctk.CTkButton(fenetre, text="Charger fichier", command=lambda: charger_fichier(context))
    bouton_charger.pack(pady=10) #affichage du bouton

    #bouton pour analyser la s√©quence
    bouton_analyser = ctk.CTkButton(fenetre, text="Lancer l‚Äôanalyse", command=lambda: analyser(context))
    bouton_analyser.pack(pady=10) #affichage du bouton

    fenetre_resultats = ctk.CTkTextbox(fenetre, height=200) #cr√©ation zone de texte pour afficher les r√©sultas
    fenetre_resultats.pack(padx=20, pady=10, fill="x") #affichage du bouton
    context["fenetre_resultats"] = fenetre_resultats #stocke la fen√®tre contenant les r√©sultats

    fenetre.mainloop()

interface()
```
